```python
#importando bibliotecas
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python

dadosFastFood = pd.read_csv("https://gist.githubusercontent.com/altinodantas/34b140583eefdb6514dc94c964b0b598/raw/a0cd494120011cc591015c667b122c0a087fa6ea/lojas.csv")

print(dadosFastFood)
```

```python
#tratando os dados
X = dadosFastFood["População"]
Y = dadosFastFood["Vendas"]

# Use reshape para transformar X e Y em arrays bidimensionais
X = np.array(X).reshape(-1, 1)
Y = np.array(Y).reshape(-1, 1)
#Criando Modelo de Regressão
lr_model = LinearRegression()
lr_model = lr_model.fit(X, Y)
```

```python
#coeficiente da Interctação
lr_model.intercept_
```

```python
#Coeficiente de angular
lr_model.coef_[0]
```

```python
#verificando o Modelo 
preds = lr_model.predict(X)
```

```python
#verificando R2
r2 = r2_score(Y, preds)
print('O valor de R2 é: {}'.format(round(r2,3),2))
```

```python
#cidade com 5mil pessoas

lr_model.predict([[5]
                  ,[30]])
```

```python
#cidade com 30mil pessoas

lr_model.predict([[30]])[0]
```
