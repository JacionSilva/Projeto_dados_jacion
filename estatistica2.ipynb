```python
dados = pd.read_csv('novo_arquivo.csv')

dados.groupby(by=['Renda']).agg(['mean'])
```

```python
renda = dados[['Renda']]
#media
media = renda.mean()
mediana = renda.median()
moda = renda.mode()
quartis1 = renda.quantile(0.25)
quartis2 = renda.quantile(0.50)
quartis3 = renda.quantile(0.75)

resultado = [media,mediana,moda,quartis1,quartis2,quartis3]
print('Média: {}'.format(str(media)))
print('Mediana: {}'.format(str(mediana)))
print('Moda: {}'.format(str(moda)))
print('Menores rendas: {}'.format(str(quartis1)))
print('Renda ideal : {}'.format(str(quartis2)))
print('Maiores rendas: {}'.format(str(quartis3)))

print('Resultado lista {}'.format(str(resultado)))
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Crie um histograma da coluna "Renda"
plt.figure(figsize=(10, 6))
plt.hist(dados['Renda'], bins=20, edgecolor='k', color='skyblue')
plt.title('Histograma da Renda')
plt.xlabel('Renda')
plt.ylabel('Frequência')
plt.grid(True)

# Exiba o histograma
plt.show()
```
